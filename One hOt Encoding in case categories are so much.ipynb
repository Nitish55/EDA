https://proceedings.mlr.press/v7/niculescu09/niculescu09.pdf In the winning solution of KDD cup 2009, Winning the KDD cup orange chalange with Ensemble selection the author limit one hot encoding to the most 10 frequent labels of the variable. This mean that they would make one binary variable for each of the 10 most frequent labels only
. This is equivalent to grouping all the other labels under one category, i.e in this case will be dropped. Thus the 10 new dummy variable indicate if one of the 10 kost frequent label is prsent(1) or not(0) for a particular observations

Note: this 10 can varies according to the sitautions, it can be 20,30 or anything accrding to situation

#### Note here e is in top 10 i.e why its getting value 0 "at", "av" are not in top 10 so its 0

```python
# Get whole set of data on dummy variables
def one_hot_encode(df,var,top_x_labels):
    # Function to create the dummy variables for most frequent variables
    # we can vary the number of most frequent labels that we can encode
    
    for labels in top_x_labels:
        df[var+"_"+labels]=np.where(df[var]==labels,1,0)
```

### For X2

```python
# Read the data again
data=pd.read_csv("https://raw.githubusercontent.com/krishnaik06/Feature-Engineering-Live-sessions/master/mercedes.csv"
                 ,usecols=['X1','X2','X3','X4','X5','X6'])
top10=[x for x in data["X2"].value_counts().sort_values(ascending=False).head(10).index]

one_hot_encode(data,"X2",top10)
data.head()
```
