### Advantages
* Easy To Use
* Not increasing feature space 
#### Disadvantages
* If some of the labels have the same count then they will be replaced by the same count and they will loose some valuable info.
* Adds somewhat arbirtatry number, and therfeore weights to the different labels, that may not be related to the predictive power.

# Target Guided Ordinal Encoding
* Ordering the labels according to the target
* Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv",usecols=["Cabin","Survived"])
df.head()
```

```python
# fill Missing value with some value
df.fillna("Missing",inplace=True)
df.head(10)
```

## Notice
* ### Here C85 means seat num 85 in C block, Lets extract the block from the dataset

```python
# Extracting only frist letter from Cabin
df["Cabin"]=[i[0] for i in df["Cabin"]]
df.head()
```

## Alternatively we can do this by

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df["Cabin"].unique()
```

```python
## finding survuval rate according to Cabin
df.groupby(["Cabin"])["Survived"].mean().sort_values()
```

* #### Here D is having the highest Survival rate

```python
ordinal_label=df.groupby(["Cabin"])["Survived"].mean().sort_values().index
ordinal_label2={k:i for i,k in enumerate(ordinal_label,0)}
ordinal_label2
```

##### Note Here D was assigned the highest rating coz it has the highest survival rate

## Mapping

```python
df["Cabin_Ordinal_Label"]=df["Cabin"].map(ordinal_label2)
df.head(8)
```

### Now we can Drop Cabin since we already treated it
# It is called Targer Guided ordinal because
##### * Based on Survival rate which we did using group by we are assigning Value to the Column "Cabin_Ordinal_Label"

#### ----------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------

# Mean Encoding Meathod
* Replacing with Mean

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean().to_dict()
mean_ordinal
```

```python
df['mean_ordinal_encode']=df['Cabin'].map(mean_ordinal)
df.head()
```

### Advantages
* It capture information within the Label
* It creates monotonic realtion betweeen the variable and the target

### Disadvantage

* it prones to overfitting

# Probablility Ratio Encoding Meathod

```python
df=pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv",usecols=["Cabin","Survived"])
df.head()
```

```python
# fill Missing value with some value
df.fillna("Missing",inplace=True)
df.head(10)
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
df.head()
```

```python
df["Cabin"].unique()
```

```python
prob_df=df.groupby(["Cabin"])["Survived"].mean()
prob_df=pd.DataFrame(prob_df)
prob_df
```

```python
# Calcultae probablity of dead
prob_df["Dead"]=1-prob_df["Survived"]
prob_df
```

### Probability Ratio Encoding

```python
prob_df["Prob_encoding"]=prob_df["Survived"]/prob_df["Dead"]
prob_df
```

```python
probability_encoding=prob_df["Prob_encoding"].to_dict()
probability_encoding
```

### Mapping with the main table DF


```python
df.head()
```

```python
df["Cabin_encoded"]=df["Cabin"].map(probability_encoding)
df.head(10)
```
